```python
provinces = [
    # Provinces (23)
    "安徽", "福建", "甘肃", "广东", "贵州", "海南", "河北", 
    "黑龙江", "河南", "湖北", "湖南", "江苏", "江西", "吉林", 
    "辽宁", "青海", "陕西", "山东", "山西", "四川", 
    "云南", "浙江", "广西", "内蒙古", "宁夏", "西藏", "新疆",
    "北京", "重庆", "上海", "天津",
    "香港", "澳门"
]
```

```python
import re

def get_provinces(text):
    provinces_found = set()  # Use a set to store unique provinces
    if isinstance(text, str):
        for province in provinces:
            if re.search(province, text):
                provinces_found.add(province)  # Add to set to ensure uniqueness
    return list(provinces_found)  # Convert back to a list

df['provinces_in_title'] = df['title'].apply(get_provinces)
df['provinces_in_content'] = df['content'].apply(get_provinces)
```
